```python
import json
import requests
import pandas as pd
```

```python
#Get subject_ids

ae_response = requests.get("https://pyhack-dot-pharmanlp-177020.uc.r.appspot.com/api/1/StudyHack/ae/subject/list")
cm_response = requests.get("https://pyhack-dot-pharmanlp-177020.uc.r.appspot.com/api/1/StudyHack/cm/subject/list")
ae_subject_ids = json.loads(ae_response.text)
cm_subject_ids = json.loads(cm_response.text)
```

```python
#Store only the subject_ids

ae_subject_ids = ae_subject_ids.get('data')
cm_subject_ids = cm_subject_ids.get('data')
```

```python
ae_data_list = []

#Iteratively fetch the url and store all the data in a list for ae

url1 = "https://pyhack-dot-pharmanlp-177020.uc.r.appspot.com/api/1/StudyHack/ae/subject/"
url2 = "/list"

for i in ae_subject_ids:
    url = url1 + str(i) + url2
    tmp = (json.loads(requests.get(url).text))
    ae_data_list.append(tmp.get('data'))
```

```python
cm_data_list = []

#Iteratively fetch the url and store all the data in a list for cm

url1 = "https://pyhack-dot-pharmanlp-177020.uc.r.appspot.com/api/1/StudyHack/cm/subject/"
url2 = "/list"

for i in cm_subject_ids:
    url = url1 + str(i) + url2
    tmp = (json.loads(requests.get(url).text))
    cm_data_list.append(tmp.get('data'))
```
